```python
data_info = pd.read_csv('../collection/artwork_data.csv', verbose=0)
data_info = data_info[['id', 'artist', 'accession_number', 'medium', 'year', 'title', 'acquisitionYear', 'thumbnailUrl']].set_index('id')
data_info = data_info[data_info.accession_number.str.startswith('P')].append(data_info[data_info.accession_number.str.startswith('T')])
data_info
```

```python
keys = data_info.accession_number.unique()
#replicating the path for each json given the start of the image id
path = {}
for key in keys:
    path[key] = os.path.join("..","collection", "artworks", key[0].lower(), key[1:4], '*.json')

path[keys[332]]
```

```python
#exploring how to retrieve all the files in the found directories
files = {}
for key in keys:
    files[key] = glob(path[key])
    
files[keys[11111]][0]
```

```python
files = [fi for file in files.values() for fi in file]
len(files)
```

```python
subjects = {}
subjects_2 = {}
import json
from tqdm import tqdm

for file in tqdm(files):
    with open(file) as json_file:
        data = json.load(json_file)
        try:
            subjects[data['acno']] = data['subjects']['children']
        except:
            continue
            #print('no subjects for', file)
        
len(subjects.values())
```

```python
subje = {}

for sub in subjects.items():
    subje[sub[0]] = [s['name'] for s in sub[1]]
    
print(subje['P01001'])
```

```python
dic = {}
for subj2, values in data.groupby('subjects_2'):
    values = values.sort_values('artist')
    dic[subj2] = {'artist':[], 'medium':[], 'creation_year':[], 'acquisition_year':[]}
    for artist, vals in values.groupby('artist'):
        dic[subj2]['artist'] += list(vals.accession_number.unique())
    values = values.sort_values('medium')
    for medium, vals in values.groupby('medium'):
        dic[subj2]['medium'] += list(vals.accession_number.unique())
    values = values.sort_values('year')
    for year, vals in values.groupby('year'):
        dic[subj2]['creation_year'] += list(vals.accession_number.unique())
    values = values.sort_values('acquisitionYear')
    for acquisitionYear, vals in values.groupby('acquisitionYear'):
        dic[subj2]['acquisition_year'] += list(vals.accession_number.unique())
```
